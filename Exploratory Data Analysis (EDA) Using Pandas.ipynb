# 4.Data Visualization

## Bar Plot – Bedrooms Count

```python
import matplotlib.pyplot as plt
ax= df['Avg. Area Number of Bedrooms'].value_counts().head(22).plot(kind='bar',title = 'Average Area Number of Bedrooms', color = 'Red')

ax.set_xlabel('no.of Bedrooms')
ax.set_ylabel("count")

plt.show()
```

#### Observation: Most houses have 3–5 bedrooms.

## Histogram – Price Distribution

```python
ax= df['Price'].plot(kind='hist', bins = 20,title = 'Room Price')

ax.set_xlabel('Price')
ax.set_ylabel('Frequency')

plt.show()
```

#### Observation: 
Prices are normally distributed, centered around $1.2M.

## Pie Chart – Bedrooms Share

```python
ax = df['Avg. Area Number of Bedrooms'].value_counts().head(20).plot(kind='pie',autopct='%1.1f%%',startangle=90,cmap='Set3',figsize=(6,6),title='Average Area Number of Bedrooms')
```

## KDE Plot – Price Density

```python
ax= df['Price'].plot(kind='kde',title = 'Room Price')
ax.set_ylabel('Density')

plt.show()
```

#### Observation:
Confirms clustering of prices near $1.2M

## Scatter Plot – Rooms vs Price

```python
df.plot(kind= 'scatter', x='Avg. Area Number of Rooms', y='Price',title= 'Avg. Area Number of Rooms vs Price')
plt.show()
```

#### Observation:
More rooms generally → Higher price.

## Scatter Plot using seaborn with Color – Rooms vs Price by House Age


```python
import seaborn as sns
sns.scatterplot(x='Avg. Area Number of Rooms', y='Price',hue='Avg. Area House Age',data = df)
plt.show()
```

#### Observation:
House age impacts price, even for similar room counts.

## Pairplot – Multivariate Analysis

```python
sns.pairplot(df, vars=['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
                       'Avg. Area Number of Bedrooms', 'Population', 'Price'])
plt.show()
```

#### Observation:

Strong correlation between income and price.

Rooms also show a positive relationship with price.
